### 1.2 Comparing Trained Model on 10 Epochs (Run 1 vs Run 2)

```python
# Setting up the plotting style and figure
sns.set(style="whitegrid")
plt.figure(figsize=(10, 15))

# Plotting Total Loss
plt.subplot(4, 1, 1)
sns.lineplot(data=df_10_epoch_train, x='epoch', y='Loss', label='Total Loss Run 1', marker='o')
sns.lineplot(data=df_10_epoch_train_run2, x='epoch', y='Loss', label='Total Loss Run 2', marker='o')
plt.title('Total Loss per Epoch for Two Runs')
plt.ylabel('Loss Value')

# Plotting SceneFlowLoss
plt.subplot(4, 1, 2)
sns.lineplot(data=df_10_epoch_train, x='epoch', y='SceneFlowLoss', label='SceneFlowLoss Run 1', marker='o')
sns.lineplot(data=df_10_epoch_train_run2, x='epoch', y='SceneFlowLoss', label='SceneFlowLoss Run 2', marker='o')
plt.title('SceneFlowLoss per Epoch for Two Runs')
plt.ylabel('SceneFlowLoss Value')

# Plotting SegLoss
plt.subplot(4, 1, 3)
sns.lineplot(data=df_10_epoch_train, x='epoch', y='SegLoss', label='SegLoss Run 1', marker='o')
sns.lineplot(data=df_10_epoch_train_run2, x='epoch', y='SegLoss', label='SegLoss Run 2', marker='o')
plt.title('SegLoss per Epoch for Two Runs')
plt.ylabel('SegLoss Value')

# Plotting TrackingLoss
plt.subplot(4, 1, 4)
sns.lineplot(data=df_10_epoch_train, x='epoch', y='TrackingLoss', label='TrackingLoss Run 1', marker='o')
sns.lineplot(data=df_10_epoch_train_run2, x='epoch', y='TrackingLoss', label='TrackingLoss Run 2', marker='o')
plt.title('TrackingLoss per Epoch for Two Runs')
plt.ylabel('TrackingLoss Value')

plt.xlabel('Epoch')
plt.tight_layout()
plt.show()
```

### Comparison of Training Results over 10 and 24 Epochs

```python

print("10 Epochs Training Results:")
print(df_10.info())
print(df_10.describe())

print("\n24 Epochs Training Results:")
print(df_24.info())
print(df_24.describe())

# Function to plot trends
def plot_trends(df, title):
    plt.figure(figsize=(14, 8))
    epoch = df['epoch']
    metrics = ['learning_rate', 'Loss', 'SceneFlowLoss', 'SegLoss', 'TrackingLoss', 'mean_train_loss']
    for metric in metrics:
        plt.plot(epoch, df[metric], label=metric)
    plt.title(title)
    plt.xlabel('Epoch')
    plt.ylabel('Value')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plotting trends for both datasets
plot_trends(df_10, 'Trends over Epochs (10 Epoch Training)')
plot_trends(df_24, 'Trends over Epochs (24 Epoch Training)')

# Comparing the best validation loss
best_val_10 = df_10.loc[df_10['best_val_loss_till_now'].idxmin()]
best_val_24 = df_24.loc[df_24['best_val_loss_till_now'].idxmin()]
```
