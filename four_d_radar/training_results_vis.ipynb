# Visualizing The Model Training Results

### This notebook takes the data aquired from training 10 and 24 epochs

## Importing Libraries

```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

## Reading the csv files

```python
# Reading csv file
_10_epoch_results = "D:\\four-d-radar-thesis\\docs\\10_epoch_training_results.csv"
data1 = pd.read_csv(_10_epoch_results)

_24_epoch_results = "D:\\four-d-radar-thesis\\docs\\24_epoch_training_results.csv"
data2 = pd.read_csv(_24_epoch_results)
```

## 1. Visualizing the Model Trained on 10 Epochs 

### Interactive Graph of Epoch vs Mean Training Loss

```python
# Setting up the figure and axis for animation
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_title('Epoch vs Mean Training Loss')
ax.set_xlabel('Epoch')
ax.set_ylabel('Mean Training Loss')
ax.grid(True)
ax.set_xlim(0, max(data1['epoch']))
ax.set_ylim(min(data1['mean_train_loss']) * 0.95, max(data1['mean_train_loss']) * 1.05)

line, = ax.plot([], [], 'bo-', label='Mean Train Loss')
ax.legend()

# Initialize the line to be empty
def init():
    line.set_data([], [])
    return line,

# Update function to draw each frame of the animation
def update(frame):
    x = data1['epoch'][:frame + 1]
    y = data1['mean_train_loss'][:frame + 1]
    line.set_data(x, y)
    return line,

# Creating the animation
ani = FuncAnimation(fig, update, frames=len(data1['epoch']), init_func=init, blit=True, repeat=False)

# Display the animation
HTML(ani.to_jshtml())
```

### Graph of Epoch vs Mean Training Loss

```python
# Setting up the figure and axis for the plot
%matplotlib inline
plt.figure(figsize=(10, 6))
plt.plot(data1['epoch'], data1['mean_train_loss'], 'bo-', label='Mean Train Loss')
plt.title('Epoch vs Mean Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Mean Training Loss')
plt.grid(True)
plt.legend()
plt.show()
```

The graph displays the mean training loss of the model over 10 epochs, showing a general downward trend indicative of learning. Initially, the loss slightly plateaus between the first and second epoch but then resumes a steadier decline, suggesting effective learning after a brief stagnation. There are small fluctuations in the later epochs, hinting at potential issues with the learning rate or parameter adjustments, but overall, the training loss has decreased from just above 0.15 to below 0.12.


### Graph of Mean Training Loss vs Best Validation Loss Over 10 Epochs

```python
# Setting up the figure and axis for the plot
plt.figure(figsize=(10, 6))
plt.plot(data1['epoch'], data1['mean_train_loss'], 'bo-', label='Mean Train Loss')
plt.plot(data1['epoch'], data1['best_val_loss_till_now'], 'ro-', label='Best Validation Loss Till Now')
plt.title('Comparison of Training and Validation Losses Over 10 Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()
plt.show()
```

### Summary of the Graph for the Model Trained on 10 Epochs

The summary of the graph is as follows:

- **Training Dynamics**: Both mean training loss and best validation loss decrease over epochs, showing that the model is learning effectively on both the training and validation datasets.
  
- **Training Loss**: It initially starts higher than the best validation loss and experiences a significant drop after the first epoch, suggesting a substantial early learning gain.

- **Best Validation Loss**: This loss decreases consistently, indicating that the model is generalizing well to new data and is not overfitting the training data.

- **Comparison Between Losses**: Initially, the best validation loss is lower than the mean training loss, which is unusual. This could be due to an easier validation set or high variance in the training loss. Later, the mean training loss falls below the best validation loss, aligning with typical expectations as the model fits the training data better.

- **Convergence**: By the final epoch, the training loss shows a slight increase, which might signal potential overfitting or simply represent variability in epoch losses. Meanwhile, the validation loss continues to decline without signs of an uptick, suggesting no immediate overfitting.

- **Graph Quality**: The graph, based on a small number of epochs, appears coarse. Increasing the number of epochs could yield a smoother curve and offer deeper insights into the training process.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df_10 = pd.read_csv(r'D:\four-d-radar-thesis\docs\10_epoch_training_results.csv')
df_24 = pd.read_csv(r'D:\four-d-radar-thesis\docs\24_epoch_training_results.csv')

# Data Inspection
print("10 Epochs Training Results:")
print(df_10.info())
print(df_10.describe())

print("\n24 Epochs Training Results:")
print(df_24.info())
print(df_24.describe())

# Function to plot trends
def plot_trends(df, title):
    plt.figure(figsize=(14, 8))
    epoch = df['epoch']
    metrics = ['learning_rate', 'Loss', 'SceneFlowLoss', 'SegLoss', 'TrackingLoss', 'mean_train_loss']
    for metric in metrics:
        plt.plot(epoch, df[metric], label=metric)
    plt.title(title)
    plt.xlabel('Epoch')
    plt.ylabel('Value')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plotting trends for both datasets
plot_trends(df_10, 'Trends over Epochs (10 Epoch Training)')
plot_trends(df_24, 'Trends over Epochs (24 Epoch Training)')

# Comparing the best validation loss
best_val_10 = df_10.loc[df_10['best_val_loss_till_now'].idxmin()]
best_val_24 = df_24.loc[df_24['best_val_loss_till_now'].idxmin()]

print("Best Validation Loss at 10 Epochs:")
print(best_val_10)
print("\nBest Validation Loss at 24 Epochs:")
print(best_val_24)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Line Plots for Loss Components
def plot_loss_components(df, title):
    sns.set(style="whitegrid")
    plt.figure(figsize=(12, 6))
    plt.title(title)
    sns.lineplot(data=df[['SceneFlowLoss', 'SegLoss', 'TrackingLoss']], dashes=False)
    plt.xlabel('Epoch')
    plt.ylabel('Loss Value')
    plt.show()

plot_loss_components(df_10, 'Loss Components Over Epochs (10 Epoch Training)')
plot_loss_components(df_24, 'Loss Components Over Epochs (24 Epoch Training)')

# Histograms for Distribution of Losses
def plot_loss_distribution(df, title):
    plt.figure(figsize=(12, 6))
    plt.title(title)
    sns.histplot(data=df[['SceneFlowLoss', 'SegLoss', 'TrackingLoss']], element="step", kde=True)
    plt.xlabel('Loss Value')
    plt.show()

plot_loss_distribution(df_10, 'Distribution of Losses (10 Epoch Training)')
plot_loss_distribution(df_24, 'Distribution of Losses (24 Epoch Training)')

# Scatter Plot with Regression for Learning Rate and Loss Correlation
def plot_lr_loss_correlation(df, title):
    sns.set(style="darkgrid")
    sns.lmplot(x='learning_rate', y='Loss', data=df, aspect=2, height=6)
    plt.title(title)
    plt.xlabel('Learning Rate')
    plt.ylabel('Loss Value')
    plt.show()

plot_lr_loss_correlation(df_10, 'Learning Rate vs Loss (10 Epoch Training)')
plot_lr_loss_correlation(df_24, 'Learning Rate vs Loss (24 Epoch Training)')

# Box Plots for Loss Spread and Outliers
def plot_loss_boxplots(df, title):
    sns.set(style="whitegrid")
    plt.figure(figsize=(12, 6))
    plt.title(title)
    sns.boxplot(data=df[['SceneFlowLoss', 'SegLoss', 'TrackingLoss']])
    plt.ylabel('Loss Value')
    plt.show()

plot_loss_boxplots(df_10, 'Loss Spread (10 Epoch Training)')
plot_loss_boxplots(df_24, 'Loss Spread (24 Epoch Training)')
```
