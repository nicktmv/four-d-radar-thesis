### Visualizing the Training Loss

```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Load the data from the specified CSV file path
data_path = "D:\\four-d-radar-thesis\\docs\\epoch_training_results.csv"
data = pd.read_csv(data_path)

# Setting up the figure and axis for animation
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_title('Epoch vs Mean Training Loss')
ax.set_xlabel('Epoch')
ax.set_ylabel('Mean Training Loss')
ax.grid(True)
ax.set_xlim(0, max(data['epoch']))
ax.set_ylim(min(data['mean_train_loss']) * 0.95, max(data['mean_train_loss']) * 1.05)

line, = ax.plot([], [], 'bo-', label='Mean Train Loss')
ax.legend()

# Initialize the line to be empty
def init():
    line.set_data([], [])
    return line,

# Update function to draw each frame of the animation
def update(frame):
    x = data['epoch'][:frame + 1]
    y = data['mean_train_loss'][:frame + 1]
    line.set_data(x, y)
    return line,

# Creating the animation
ani = FuncAnimation(fig, update, frames=len(data['epoch']), init_func=init, blit=True, repeat=False)

# Display the animation
HTML(ani.to_jshtml())
```
