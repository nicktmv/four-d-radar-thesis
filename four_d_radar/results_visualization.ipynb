# Visualizing The Model Training Results

### This notebook takes the results aqquired from the model trained on 8, 10 and 24 epochs across different configurations defined in the config.yaml file 

## Importing Libraries

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading the Data

```python
epoch_8_root = "/four-d-radar-thesis/docs/results/epoch-8"
epoch_10_root = "/four-d-radar-thesis/docs/results/epoch-10"

# 8 Epochs Run 1 (pretrain_epochs = 2)
df_8_epoch_loss_run1 = pd.read_csv(f'{epoch_8_root}/8_epoch_loss_run1.csv')
df_8_epoch_segmentation_run1 = pd.read_csv(f'{epoch_8_root}/8_epoch_segmentation_run1.csv')
df_8_epoch_scene_flow_run1 = pd.read_csv(f'{epoch_8_root}/8_epoch_scene_flow_run1.csv')

# 8 Epochs Run 2 (pretrain_epochs = 16)
df_8_epoch_loss_run2 = pd.read_csv(f'{epoch_8_root}/8_epoch_loss_run2.csv')
df_8_epoch_segmentation_run2 = pd.read_csv(f'{epoch_8_root}/8_epoch_segmentation_run2.csv')
df_8_epoch_scene_flow_run2 = pd.read_csv(f'{epoch_8_root}/8_epoch_scene_flow_run2.csv')

# 10 Epochs Run 1 (pretrain_epochs = 16)
df_10_epoch_loss_run1 = pd.read_csv(f'{epoch_10_root}/10_epoch_loss_run1.csv')

# 10 Epochs Run 2 (pretrain_epochs = 2)
df_10_epoch_loss_run2 = pd.read_csv(f'{epoch_10_root}/10_epoch_loss_run2.csv')
df_10_epoch_segmentation_run2 = pd.read_csv(f'{epoch_10_root}/10_epoch_segmentation_run2.csv')
df_10_epoch_scene_flow_run2 = pd.read_csv(f'{epoch_10_root}/10_epoch_scene_flow_run2.csv')

# 24 Epochs Run 1 (pretrain_epochs = 2)
#df_24_epoch_train = pd.read_csv(f'{root}/24_epoch_training_results.csv')

# comparison of segmentation results
#df_segmentation_comparison = pd.read_csv(f'{root}/segmentation_comparison.csv')
```

### 1.2 Comparing Trained Model on 8 Epochs (Run 1 vs Run 2)
Run 1 (pretrained_epochs = 2)<br>
Run 2 (pretrained_epochs = 16)

```python
# Setting up the plotting style and figure
sns.set(style="whitegrid")
plt.figure(figsize=(10, 15))

# Plotting Total Loss
plt.subplot(4, 1, 1)
sns.lineplot(data=df_8_epoch_loss_run1, x='epoch', y='Loss', label='pretrained_epochs: 2', marker='o')
sns.lineplot(data=df_8_epoch_loss_run2, x='epoch', y='Loss', label='pretrained_epochs: 16', marker='o')
plt.title('Total Loss per 8 Epochs for Two Runs')
plt.ylabel('Loss Value')

# Plotting SceneFlowLoss
plt.subplot(4, 1, 2)
sns.lineplot(data=df_8_epoch_loss_run1, x='epoch', y='SceneFlowLoss', label='pretrained_epochs: 2', marker='o')
sns.lineplot(data=df_8_epoch_loss_run2, x='epoch', y='SceneFlowLoss', label='pretrained_epochs: 16', marker='o')
plt.title('SceneFlowLoss per 8 Epochs for Two Runs')
plt.ylabel('SceneFlowLoss Value')

# Plotting SegLoss
plt.subplot(4, 1, 3)
sns.lineplot(data=df_8_epoch_loss_run1, x='epoch', y='SegLoss', label='pretrained_epochs: 2', marker='o')
sns.lineplot(data=df_8_epoch_loss_run2, x='epoch', y='SegLoss', label='pretrained_epochs: 16', marker='o')
plt.title('SegLoss per 8 Epochs for Two Runs')
plt.ylabel('SegLoss Value')

# Plotting TrackingLoss
plt.subplot(4, 1, 4)
sns.lineplot(data=df_8_epoch_loss_run1, x='epoch', y='TrackingLoss', label='pretrained_epochs: 2', marker='o')
sns.lineplot(data=df_8_epoch_loss_run2, x='epoch', y='TrackingLoss', label='pretrained_epochs: 16', marker='o')
plt.title('TrackingLoss per 8 Epochs for Two Runs')
plt.ylabel('TrackingLoss Value')

plt.xlabel('Epoch')
plt.tight_layout()
plt.show()
```

### 2.3 Segmentation Comparison

#### 2.3.1 Comparing 8 Epochs Run 1 vs Run 2
Run 1 (pretrained_epochs = 2)<br>
Run 2 (pretrained_epochs = 16)

```python
# Compute average metrics for run 1
avg_acc_run1 = df_8_epoch_segmentation_run1['acc'].mean()
avg_miou_run1 = df_8_epoch_segmentation_run1['miou'].mean()
avg_sen_run1 = df_8_epoch_segmentation_run1['sen'].mean()
# Compute average metrics for run 2
avg_acc_run2 = df_8_epoch_segmentation_run2['acc'].mean()
avg_miou_run2 = df_8_epoch_segmentation_run2['miou'].mean()
avg_sen_run2 = df_8_epoch_segmentation_run2['sen'].mean()

# Creating a summary dataframe
summary_df = pd.DataFrame({
    'Metric': ['Accuracy', 'MIoU', 'Sensitivity'],
    'Run 1': [avg_acc_run1, avg_miou_run1, avg_sen_run1],
    'Run 2': [avg_acc_run2, avg_miou_run2, avg_sen_run2],
    'Better': [
        "Run 1" if avg_acc_run1 > avg_acc_run2 else "Run 2",
        "Run 1" if avg_miou_run1 > avg_miou_run2 else "Run 2",
        "Run 1" if avg_sen_run1 > avg_sen_run2 else "Run 2"
    ]
})

summary_df
```

## 3. Scene Flow Comparison

### 3.1 Comparing 8 Epochs Run 1 vs Run 2
Run 1 (pretrained_epochs = 2)<br>
Run 2 (pretrained_epochs = 16)

```python
# Compute average metrics for scene flow run 1
avg_rne_run1 = df_8_epoch_scene_flow_run1['rne'].mean()
avg_sas_run1 = df_8_epoch_scene_flow_run1['sas'].mean()
avg_epe_run1 = df_8_epoch_scene_flow_run1['epe'].mean()
avg_ras_run1 = df_8_epoch_scene_flow_run1['ras'].mean()  # Added RAS

# Compute average metrics for scene flow run 2
avg_rne_run2 = df_8_epoch_scene_flow_run2['rne'].mean()
avg_sas_run2 = df_8_epoch_scene_flow_run2['sas'].mean()
avg_epe_run2 = df_8_epoch_scene_flow_run2['epe'].mean()
avg_ras_run2 = df_8_epoch_scene_flow_run2['ras'].mean()  # Added RAS

# Creating a summary dataframe for scene flow comparison
scene_flow_summary_df = pd.DataFrame({
    'Metric': ['RNE', 'SAS', 'EPE', 'RAS'],
    'Run 1': [avg_rne_run1, avg_sas_run1, avg_epe_run1, avg_ras_run1],
    'Run 2': [avg_rne_run2, avg_sas_run2, avg_epe_run2, avg_ras_run2],
    'Better': [
        "Run 1" if avg_rne_run1 < avg_rne_run2 else "Run 2",  # Note: Lower RNE and EPE are better
        "Run 1" if avg_sas_run1 > avg_sas_run2 else "Run 2",
        "Run 1" if avg_epe_run1 < avg_epe_run2 else "Run 2",
        "Run 1" if avg_ras_run1 > avg_ras_run2 else "Run 2"   # Higher RAS is better
    ]
})

scene_flow_summary_df
```
